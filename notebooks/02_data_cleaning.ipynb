```python
import pandas as pd
import numpy as np
```

```python
churn = pd.read_csv("../data/BankChurners.csv")

fraud = pd.read_csv("../data/creditcard.csv")

marketing = pd.read_csv(
    "../data/marketing_campaign.csv",
    sep='\t'
)


retail = pd.read_excel(
    "../data/online_retail_II.xlsx"
)


engagement = pd.read_csv(
    "../data/engagement.csv"
)


merchant = pd.read_csv(
    "../data/merchant_master.csv"
)
```

```python
datasets = {
    'Churn': churn,
    'Fraud': fraud,
    'Marketing': marketing,
    'Retail': retail,
    'Engagement': engagement,
    'Merchant': merchant
}

for name, df in datasets.items():
    print("="*50)
    print(name)
    print("Shape:", df.shape)
    print()
```

```python
for name,df in datasets.items():

    print("="*50)

    print(name)

    print(df.isnull().sum())

    print()
```

```python
for name,df in datasets.items():

    print("="*50)

    print(name)

    print("Duplicates :",df.duplicated().sum())

    print()
```

```python
for df in datasets.values():

    df.columns = (

        df.columns
        .str.lower()
        .str.strip()
        .str.replace(" ","_")

    )
```

```python
churn.to_csv(
'../data/processed/churn_clean.csv',
index=False
)

fraud.to_csv(
'../data/processed/fraud_clean.csv',
index=False
)

marketing.to_csv(
'../data/processed/marketing_clean.csv',
index=False
)

retail.to_csv(
'../data/processed/retail_clean.csv',
index=False
)

engagement.to_csv(
'../data/processed/engagement_clean.csv',
index=False
)

merchant.to_csv(
'../data/processed/merchant_clean.csv',
index=False
)
```
