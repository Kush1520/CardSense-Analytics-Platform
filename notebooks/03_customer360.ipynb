```python
import pandas as pd
import numpy as np

churn = pd.read_csv(
    "../data/processed/churn_clean.csv"
)

marketing = pd.read_csv(
    "../data/processed/marketing_clean.csv"
)

engagement = pd.read_csv(
    "../data/processed/engagement_clean.csv"
)

retail = pd.read_csv(
    "../data/processed/retail_clean.csv"
)

merchant = pd.read_csv(
    "../data/processed/merchant_clean.csv"
)
```

```python
print(churn.columns)
print()

print(marketing.columns)
print()

print(engagement.columns)
```

```python
churn_df = churn[[

'clientnum',

'customer_age',

'gender',

'credit_limit',

'total_trans_amt',

'total_trans_ct',

'avg_utilization_ratio',

'attrition_flag'

]]
```

```python
churn_df.columns=[

'customer_id',

'age',

'gender',

'credit_limit',

'total_spend',

'transaction_count',

'utilization',

'churn_flag'

]
```

```python
marketing_df = marketing[[

'id',

'income',

'education',

'response'

]]
```

```python
marketing_df.columns=[

'customer_id',

'income',

'education',

'campaign_response'

]
```

```python
engagement_df = engagement.copy()


engagement_df['engagement_score']=(

engagement_df['app_logins']*0.4+

engagement_df['email_clicks']*0.3+

engagement_df['offer_acceptance']*20-

engagement_df['complaints']*5

)


engagement_df = engagement_df[[

'customer_id',

'engagement_score',

'offer_acceptance'

]]
```

```python
customer360 = churn_df.merge(

    engagement_df,

    on='customer_id',

    how='left'

)
```
