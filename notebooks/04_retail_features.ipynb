```python
import pandas as pd
import numpy as np
```

```python
retail = pd.read_csv(
    "../data/processed/retail_clean.csv"
)

customer360 = pd.read_csv(
    "../data/customer360.csv"
)
```

```python
retail.head()
```

```python
retail.columns
```

```python
retail['sales']=retail['quantity']*retail['price']
```

```python
retail_features = retail.groupby(

'customer_id'

).agg(

frequency=('invoice','nunique'),

monetary=('sales','sum'),

avg_order_value=('sales','mean')

).reset_index()
```

```python
retail_features.head()
```

```python
customer360 = customer360.merge(

retail_features,

on='customer_id',

how='left'

)
```

```python
customer360.fillna(0,inplace=True)
```

```python
customer360.to_csv(

'../data/customer360_enriched.csv',

index=False

)
```
