```python
import pandas as pd
import numpy as np
import random
```

```python
merchant_names = [

'Amazon',
'Walmart',
'Target',
'Starbucks',
'Shell',
'Costco',
'Best Buy',
'Hilton',
'Marriott',
'Delta',
'United',
'Uber',
'Netflix',
'Apple',
'CVS'

]


categories = [

'Shopping',
'Dining',
'Travel',
'Fuel',
'Healthcare',
'Electronics',
'Entertainment',
'Hotels',
'Airlines',
'Groceries'

]


countries = [

'USA',
'Canada',
'UK',
'Australia',
'Singapore'

]
```

```python
n_merchants = 5000


merchant_df = pd.DataFrame({

'merchant_id':[f'M{i:05}' for i in range(1,n_merchants+1)],


'merchant_name':

np.random.choice(
merchant_names,
n_merchants
),


'category':

np.random.choice(
categories,
n_merchants
),



'country':

np.random.choice(
countries,
n_merchants
),



'risk_score':

np.round(

np.random.uniform(
0,
1,
n_merchants
),

2

)

})



merchant_df.head()
```

```python
merchant_df.to_csv(

'../data/merchant_master.csv',

index=False

)
```
